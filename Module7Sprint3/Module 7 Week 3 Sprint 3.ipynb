```python
#Import pandas to build a DataFrame and the import DataFrame
import pandas as pd
from pandas import DataFrame
#Import csv to read from the csv formatted text file
import csv
#Import numpy to build an array
import numpy as np
#Import matplotlib for plotting the pie charts
import matplotlib.pyplot as plt
#Opening the text file
path="Corona_Data.txt"
corona_stat_file=open(path, mode='r', newline='')
#Reading from the text file
reader=csv.reader(corona_stat_file, delimiter=',')
print(reader)
headings=next(reader)
print(headings)
#Creating arrays 
data=[]
Country=[]
Country_NewCases=[]
Country_TotalCases=[]
Country_NewDeaths=[]
Country_TotalDeaths=[]
#Adding the total reading of cases and deaths to the created arrays
for row in reader:
    if row[3]=="2020-04-23T00:00:00.000Z":
       Country.append(row[2])
       Country_NewCases.append(row[4])
       Country_TotalCases.append(row[5])
       Country_NewDeaths.append(row[6])
       Country_TotalDeaths.append(row[7])
#Creating a dtaframe for the data
Cases_Deaths=np.array([Country_TotalCases,Country_TotalDeaths])
Cases_Death=np.reshape(Cases_Deaths, (83,2))
case_graph = pd.DataFrame(data=Cases_Death, columns=["Total Cases","Total Deaths"] , index=Country) 
print(case_graph)
#Creating a line chart with all the data
def Line_Chart():
   linegraph=plt.plot(Country,Country_TotalCases)
   linegraph=plt.plot(Country,Country_TotalDeaths)
   linegraph=plt.xlabel('Amount of tweets about topic')
   linegraph=plt.ylabel('Dlinegraph=ate topic was tweeted about')
   linegraphpl=plt.title('Graph of Total Cases and Deaths per a Country')
Line_Chart()
#Creating a pie chart with all the data
def Pie_Chart():
   colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','green','grey','white','orange','red','pink']
   plt.pie(Country_TotalDeaths, colors=colors, autopct='%0.0f%%', shadow=False, startangle=140)
   plt.axis('equal')
   patches, texts = plt.pie(Country_TotalDeaths, colors=colors, shadow=False, startangle=20)
   plt.legend(patches, Country)
   plt.show()  
Pie_Chart()   
#Creating a pie chart with all the data
def Pie_Chart2():
   colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','green','grey','white','orange','red','pink'] 
   plt.pie(Country_TotalCases, colors=colors, autopct='%0.0f%%', shadow=False, startangle=140)
   plt.axis('equal')
   patches, texts = plt.pie(Country_TotalCases, colors=colors, shadow=False, startangle=20)
   plt.legend(patches, Country)
   plt.show() 
Pie_Chart2()   
```
